Libraries Imported

```python
from bs4 import BeautifulSoup
import requests
```

URLs Requests

```python
URLs = 'http://www.ftu.ac.th/2019/index.php/th/'
Re = requests.get(URLs)
```

```python
Re.text
```

BeautifulSoup

```python
Soup = BeautifulSoup(Re.content, 'html.parser')
```

```python
def print_hyperlinks():
    for links in Soup.find_all('a'):
        print(links)
```

```python
print_hyperlinks()
```
